b. Top 3 prodi apa saja yang paling banyak diminati oleh para fresh graduate? Untuk mengetahui top 3 program studi yang paling banyak diminati oleh para fresh graduate, kita perlu mengambil data dari tingkat pendidikan S1 (Strata Satu). Berdasarkan data penerimaan mahasiswa tahun 2023, top 3 program studi yang paling banyak diminati oleh para fresh graduate adalah:
* Akuntansi (50 mahasiswa)
* Manajemen (43 mahasiswa)
* Ilmu Komputer (40 mahasiswa)

```python
#Data penerimaan mahasiswa tahun 2023
mahasiswa_2023 = {
    'Tingkat Pendidikan': ['S2', 'S1', 'S1', 'S1', 'S1', 'S1', 'D3', 'S1', 'S1', 'S1'],
    'Prodi': ['Ilmu Komputer', 'Sistem Informasi', 'Informatika', 'Sains Data', 'Manajemen',
              'Bisnis Digital', 'Perhotelan', 'Sastra Inggris', 'Akuntansi', 'Multimedia'],
    'Jumlah': [40, 36, 38, 15, 43, 23, 21, 30, 50, 20]
}

# Membuat DataFrame dari data
df = pd.DataFrame(mahasiswa_2023)

# Mengelompokkan data berdasarkan program studi dan menghitung total penerimaan mahasiswa untuk setiap program studi
total_penerimaan = df.groupby('Prodi')['Jumlah'].sum().reset_index()

# Membuat bar graph
plt.figure(figsize=(10, 6))
plt.bar(total_penerimaan['Prodi'], total_penerimaan['Jumlah'], color='skyblue')
plt.title('Total Penerimaan Mahasiswa per Program Studi Tahun 2023')
plt.xlabel('Program Studi')
plt.ylabel('Jumlah Penerimaan Mahasiswa')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Menampilkan top 3 prodi yang paling diminati oleh para fresh graduate
top_3_prodi = total_penerimaan.sort_values(by='Jumlah', ascending=False).head(3)
print("Top 3 Prodi yang Paling Diminati oleh Para Fresh Graduate:")
print(top_3_prodi)
```

Untuk masalah ini, kami akan memilih diagram batang (bar chart) karena ini akan memudahkan untuk membandingkan jumlah penjualan atau popularitas setiap jenis bunga. b. 10 bunga apa saja yang anda sarankan untuk dihapus dari jenis bunga yang dijual? Untuk mengetahui 10 bunga apa saja yang harus dihapus untuk mencegah kebangkrutan pada toko tersebut, kami menyajikan 10 jenis bunga dengan penjualan ter-rendah berdasarkan data :
* Bunga Matahari(35)
* Matahari Kuning(30)
* Lily Putih(28)
* Bunga Dahlia(25)
* Melati(24)
* Anggrek Bulan(20)
* Tulip Kuning(17)
* Bunga Krisan(15)
* Bunga Bakung(10)
* Bunga Matahari Kecil(5)

```python
# Data penjualan bunga
data = {
    'Jenis Bunga': ['Mawar Merah', 'Lily Putih', 'Tulip Kuning', 'Melati', 'Anggrek Bulan',
                    'Matahari Kuning', 'Bunga Bakung', 'Bunga Matahari', 'Lavender Ungu',
                    'Bunga Krisan', 'Bunga Dahlia', 'Bunga Aster', 'Bunga Kamboja',
                    'Bunga Sakura', 'Bunga Anggrek', 'Bunga Lili', 'Bunga Anggrek Hitam',
                    'Bunga Krokus', 'Bunga Daisy', 'Bunga Matahari Kecil'],
    'Penjualan': [80, 28, 17, 24, 20, 30, 10, 35, 40, 15, 25, 45, 50, 70, 75, 60, 55, 50, 65, 5]
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Mengurutkan data berdasarkan jumlah penjualan
df_sorted = df.sort_values(by='Penjualan', ascending=False)

# Memilih 10 bunga yang akan dihapus dari jenis bunga yang dijual
bunga_hapus = df_sorted.tail(10)

# Membuat bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['Jenis Bunga'], df['Penjualan'], color='lightblue')
plt.title('Penjualan Bunga di Flowerensita')
plt.xlabel('Jenis Bunga')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Menampilkan 10 jenis bunga yang disarankan dihapus dari menu
print("Bunga yang Disarankan untuk Dihapus dari Jenis Bunga yang Dijual:")
print(bunga_hapus)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Soal 2
1.	Visualisasikan dan olah data diatas

```python


# Data jumlah tidur malam dan produktivitas kerja
data = {
    'Jumlah Tidur (jam)': [7, 8, 9, 5, 4, 6, 7, 8, 4, 2, 1, 2, 3, 4, 5, 4, 3, 2, 6],
    'Produktivitas Kerja': [7, 9, 7, 6, 5, 4, 5, 7, 8, 9, 8, 7, 6, 5, 4, 5, 6, 4, 3]
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Visualisasi data menggunakan scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Jumlah Tidur (jam)'], df['Produktivitas Kerja'], color='blue')
plt.title('Hubungan antara Jumlah Tidur Malam dan Produktivitas Kerja')
plt.xlabel('Jumlah Tidur (jam)')
plt.ylabel('Produktivitas Kerja')
plt.grid(True)
plt.show()
```

Soal 2
2. Apakah ada hubungan antara jumlah tidur malam dan produktivitas kerja?

  Untuk mengetahui hubungan antara kedua variabel tersebut, kita bisa melihat pola umum pada scatter plot. Jika ada pola naik/turun yang jelas, itu menunjukkan adanya hubungan. Dari visualisasi, tampaknya ada tren positif, yaitu semakin banyak tidur malam, semakin tinggi produktivitas kerja.

Soal 2
3. Berapa rata-rata jumlah tidur malam yang didapatkan individu dalam data ini?

```python
# Rata-rata jumlah tidur malam yang didapatkan individu dalam data ini
rata_rata_tidur = df['Jumlah Tidur (jam)'].mean()
print("Rata-rata jumlah tidur malam:", rata_rata_tidur, "Jam")
```

Soal 2
4.	Apakah ada korelasi antara jumlah tidur malam dan produktivitas kerja? Jika ada, seberapa kuat korelasinya?

```python

# Korelasi antara jumlah tidur malam dan produktivitas kerja
korelasi = df['Jumlah Tidur (jam)'].corr(df['Produktivitas Kerja'])
print("Korelasi antara jumlah tidur malam dan produktivitas kerja:", korelasi)
# Jika nilai korelasi mendekati 1, itu menunjukkan hubungan positif yang kuat.
# Jika nilai korelasi mendekati -1, itu menunjukkan hubungan negatif yang kuat.
```

Soal 2
5.	Bagaimana distribusi produktivitas kerja di antara individu dalam data ini?

```python
# Distribusi produktivitas kerja di antara individu dalam data ini
plt.figure(figsize=(8, 6))
plt.hist(df['Produktivitas Kerja'], bins=range(11), color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribusi Produktivitas Kerja')
plt.xlabel('Produktivitas Kerja')
plt.ylabel('Frekuensi')
plt.xticks(range(11))
plt.grid(True, axis='y')
plt.show()
```

  Pola yang menonjol dalam hubungan antara jumlah tidur malam dan produktivitas kerja. Dari scatter plot sebelumnya, kita dapat melihat bahwa produktivitas kerja cenderung lebih tinggi saat jumlah tidur malam lebih dari 7 jam. Namun, ini hanya berdasarkan pada data yang tersedia.

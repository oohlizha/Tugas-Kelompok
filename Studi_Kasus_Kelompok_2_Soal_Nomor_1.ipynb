```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Soal 1
1. Memvisualisasikan jumlah produk yang terjual dan pendapatan untuk setiap cabang


```python

# Data penjualan dan pendapatan dari beberapa cabang restoran
data = {
    'Menu': ['Ayam Goreng', 'Nasi Kuning', 'Bubur Ayam', 'Tahu Goreng', 'Pecel Lele', 'Gado-gado',
               'Mie Goreng', 'Sop Buntut', 'Bakso Goreng', 'Rawon', 'Nasi Goreng', 'Mie Ayam',
               'Sate Ayam', 'Steak', 'Es Teh', 'Es Campur', 'Soto Ayam', 'Ikan Bakar', 'Capcay'],
    'Jumlah Produk Terjual': [130, 90, 110, 85, 70, 120, 100, 75, 95, 80, 120, 150, 80, 100, 200, 180, 90, 70, 110]
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Mengurutkan data berdasarkan jumlah penjualan
df_sorted = df.sort_values(by='Jumlah Produk Terjual', ascending=False)

# Membuat bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['Menu'], df['Jumlah Produk Terjual'], color='lightpink')
plt.title('Penjualan Produk per Cabang Restoran')
plt.xlabel('Menu Restoran')
plt.ylabel('Jumlah Produk Terjual')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

2. Memperhatikan perbedaan performa antara jenis usaha "Makanan Cepat Saji" dan "Restoran Keluarga".


```python
# Data penjualan dan pendapatan dari beberapa cabang restoran
data = {
    'Menu': ['Ayam Goreng', 'Nasi Kuning', 'Bubur Ayam', 'Tahu Goreng', 'Pecel Lele', 'Gado-gado',
               'Mie Goreng', 'Sop Buntut', 'Bakso Goreng', 'Rawon', 'Nasi Goreng', 'Mie Ayam',
               'Sate Ayam', 'Steak', 'Es Teh', 'Es Campur', 'Soto Ayam', 'Ikan Bakar', 'Capcay'],
    'Jenis Usaha': ['Makanan Cepat Saji', 'Makanan Cepat Saji', 'Makanan Cepat Saji', 'Makanan Cepat Saji',
                    'Makanan Cepat Saji', 'Restoran Keluarga', 'Makanan Cepat Saji', 'Restoran Keluarga',
                    'Makanan Cepat Saji', 'Restoran Keluarga', 'Makanan Cepat Saji', 'Makanan Cepat Saji',
                    'Makanan Cepat Saji', 'Restoran Keluarga', 'Minuman', 'Minuman', 'Makanan Cepat Saji',
                    'Restoran Keluarga', 'Makanan Cepat Saji'],
    'Jumlah Produk Terjual': [130, 90, 110, 85, 70, 120, 100, 75, 95, 80, 120, 150, 80, 100, 200, 180, 90, 70, 110],
    'Pendapatan': [np.random.randint(1000, 5000) for _ in range(19)]  # Random pendapatan untuk setiap cabang
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Membuat bar graph
plt.figure(figsize=(12, 6))
for jenis_usaha, group in df.groupby('Jenis Usaha'):
    plt.bar(group['Menu'], group['Jumlah Produk Terjual'], label=jenis_usaha)

plt.title('Penjualan Produk per Cabang Restoran')
plt.xlabel('Menu Restoran')
plt.ylabel('Jumlah Produk Terjual')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()

# Memperhatikan perbedaan performa antara jenis usaha "Makanan Cepat Saji" dan "Restoran Keluarga"
performa_mcs = df[df['Jenis Usaha'] == 'Makanan Cepat Saji']['Jumlah Produk Terjual'].sum()
performa_rk = df[df['Jenis Usaha'] == 'Restoran Keluarga']['Jumlah Produk Terjual'].sum()

print("Perbedaan Performa Antara Jenis Usaha 'Makanan Cepat Saji' dan 'Restoran Keluarga':")
print(f"Total produk terjual di Makanan Cepat Saji: {performa_mcs}")
print(f"Total produk terjual di Restoran Keluarga: {performa_rk}")

# Penjelasan mengapa menggunakan grafik ini
print("\nPenjelasan mengapa menggunakan grafik ini:")
print("Grafik ini dipilih karena diagram batang memungkinkan kita untuk dengan jelas membandingkan jumlah produk yang terjual dan pendapatan dari setiap cabang restoran.")
```
